```python
from nvd3 import pieChart
from nvd3 import discreteBarChart
```

```python
# Open File to write the D3 Graph
output_file = open('test-nvd3.html', 'w')

type = 'pieChart'
chart = pieChart(name=type, color_category='category20c', height=450, width=450)
chart.set_containerheader("\n\n<h2>" + type + "</h2>\n\n")

xdata = ["Orange", "Banana", "Pear", "Kiwi", "Apple", "Strawberry", "Pineapple"]
ydata = [3, 4, 0, 1, 5, 7, 3]

extra_serie = {"tooltip": {"y_start": "", "y_end": " cal"}}
chart.add_serie(y=ydata, x=xdata, extra=extra_serie)
chart.buildhtml()
output_file.write(chart.htmlcontent)
output_file.close()
```

```python
output_file = open('test-nvd3.html', 'a')
chart = discreteBarChart(name='discreteBarChart', height=400, width=400)
chart.set_containerheader("\n\n<h2>"+"Bar Chart"+"</h>\n\n")

xdata = ["A", "B", "C", "D", "E", "F"]
ydata = [3, 4, 0, -3, 5, 7]

chart.add_serie(y=ydata, x=xdata)
chart.buildhtml()
output_file.write(chart.htmlcontent)
# close Html file
output_file.close()
```

```python
def all_calc(est_currency,min_p,r_p,max_p):
    p3_prob=(min_p+r_p+max_p)/3
    pert_prob=(min_p+4*r_p+max_p)/6
    
    min_value=est_currency*min_p/100
    max_value=est_currency*max_p/100
    avg_value=(min_value+max_value)/2
    real_value=est_currency*r_p/100
    
    p3_value=(min_value+real_value+max_value)/3
    pert_value=(min_value+4*real_value+max_value)/6
    return p3_prob,pert_prob,min_value,max_value,avg_value,real_value,p3_value,pert_value

all_calc(250000,40,75,85)
df_S[['PARAM NAME','MIN PROB ADJUSTED VALUE','MAX PROB ADJUSTED VALUE','AVERAGE PROB ADJUSTED VALUE',
     'REALISTIC PROB ADJUSTED VALUE','3 POINT BASED PROB ADJUSTED VALUE','PERT BASED PROB ADJUSTED VALUE']
    ]
```

```python
df_S=pd.read_excel("../data/Strength AnalysisData.xlsx")
df_O=pd.read_excel("../data/Opportunity Analysis Data.xlsx")
df_W=pd.read_excel("../data/Weakness Analysis Data.xlsx")
df_T=pd.read_excel("../data/Threat Analysis Data.xlsx")

df_all=pd.read_excel("../data/All Four Factors.xlsx")

df_con=pd.concat([df_S,df_O,df_W,df_T])
df_con
```

```python

df_con=df_con[['CATEGORY','FACTOR TYPE','PARAM NAME','MIN PROB ADJUSTED VALUE','MAX PROB ADJUSTED VALUE', 'AVERAGE PROB ADJUSTED VALUE',
'REALISTIC PROB ADJUSTED VALUE', '3 POINT BASED PROB ADJUSTED VALUE','PERT BASED PROB ADJUSTED VALUE']]
df_con.set_axis(['Cat','PN','Parameter','Min','Max','Avg','Realistic','3PT','PERT'], axis='columns', inplace=True)
df_con.reset_index(drop=True)
df_con.index = pd.RangeIndex(start=0, stop=len(df_con), step=1)

df_conMod=pd.DataFrame(np.zeros((len(df_con)*6,5)),columns=['Category','Effect','Parameter','Data Points','Adjusted Value'])
for i in range(len(df_con)):
    df_conMod.loc[6*i:(6*i+6),'Parameter']=df_con.loc[i,'Parameter']
    df_conMod.loc[6*i:(6*i+6),'Category']=df_con.loc[i,'Cat']
    df_conMod.loc[6*i:(6*i+6),'Effect']=df_con.loc[i,'PN']
    
    for j in range(6):
        df_conMod.loc[6*i+j,'Data Points']=['Min','Max','Avg','Realistic','3PT','PERT'][j]
df_conMod
for i in range(len(df_conMod)):
    param=df_conMod.loc[i,"Parameter"]
    dp=df_conMod.loc[i,'Data Points']
    df_conMod.loc[i,'Adjusted Value']=df_con[df_con["Parameter"]==param][dp].tolist()[0]

for i in range(len(df_conMod)):
    if df_conMod.loc[i,"Category"]=="Strength" or df_conMod.loc[i,"Category"]=="Weakness":
        df_conMod.loc[i,"Category"]="Internal"
    else:
        df_conMod.loc[i,"Category"]="External"
df_conMod
```

```python
from nvd3 import multiBarChart
from IPython.core.display import HTML

output_file = open('threat.html', 'w')
xdata=['Min','Max','Avg','Realistic','3PT','PERT']

chart=multiBarChart(width=800, height=400, x_axis_format=None)
chart.set_containerheader("\n\n<h2>" + "Threat: Adjusted Value in K" + "</h2>\n\n")
for i in range(len(df_T)):
    tempA=df_T.loc[i,['MIN PROB ADJUSTED VALUE','MAX PROB ADJUSTED VALUE', 'AVERAGE PROB ADJUSTED VALUE',
'REALISTIC PROB ADJUSTED VALUE', '3 POINT BASED PROB ADJUSTED VALUE','PERT BASED PROB ADJUSTED VALUE']]
    tempA=[int(i/1000) for i in list(tempA)]

    chart.add_serie(name=df_T.loc[i,"PARAM NAME"],y=tempA,x=xdata)
chart.buildhtml()
chart.create_x_axis("Data Points",label="Data Points")
chart.create_y_axis("Adjusted Value",label="Adjusted Value in Thousands")
output_file.write(chart.htmlcontent)
output_file.close()
# HTML(chart.htmlcontent)
```

```python
df_SMod=df_S[['PARAM NAME','MIN PROB ADJUSTED VALUE','MAX PROB ADJUSTED VALUE', 'AVERAGE PROB ADJUSTED VALUE',
'REALISTIC PROB ADJUSTED VALUE', '3 POINT BASED PROB ADJUSTED VALUE','PERT BASED PROB ADJUSTED VALUE']]
df_SMod.set_axis(['Parameter','Min','Max','Avg','Realistic','3PT','PERT'], axis='columns', inplace=True)
df_SMod

df_vis=pd.DataFrame(np.zeros((18,3)),columns=['Parameter','Data Points','Adjusted Value'])
for i in range(3):
    df_vis.loc[6*i:(6*i+6),'Parameter']=df_SMod.loc[i,'Parameter']
    for j in range(6):
        df_vis.loc[6*i+j,'Data Points']=['Min','Max','Avg','Realistic','3PT','PERT'][j]

for i in range(len(df_vis)):
    param=df_vis.loc[i,"Parameter"]
    dp=df_vis.loc[i,'Data Points']
    df_vis.loc[i,'Adjusted Value']=df_SMod[df_SMod["Parameter"]==param][dp].tolist()[0]
df_vis
chart=alt.Chart(df_vis).mark_bar().encode(
x='Data Points',
y='Adjusted Value',
color='Parameter').properties(
    width=250,
    height=250
).interactive()
display(chart)

# display((chart | chart1))
```
